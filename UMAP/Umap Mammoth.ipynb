## UMAP

```python
import pandas as pd
import numpy as np
import umap.umap_ as umap
%matplotlib inline

import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib as mpl
#For Tables:
from IPython.display import display
from IPython.display import Latex
pd.set_option('display.max_columns', 500)
from tqdm import tnrange, tqdm_notebook
```

```python
mammoth = pd.read_csv('mammut_umap.csv')
```

```python
mammoth = mammoth.sample(30000) #porque tiene mas de 400k datos
```

```python
mammoth.dropna(inplace=True)
```

```python
#plot the mammoth
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(90,72))

ax = fig.add_subplot(111, projection='3d')
# ax.set_aspect('equal')

ax.set_axis_off()
ax.scatter(mammoth['x'], mammoth['y'], mammoth['z'], s=20,c='black')
ax.view_init(0, -170)

plt.show()
```

```python
reducer = umap.UMAP(random_state=42,
                    n_components=2,
                    n_neighbors=30,
                    metric='euclidean')

reducer.fit(mammoth)
```

```python
plt.figure(figsize=(40,40),facecolor='w')

plt.axis('off')
plt.scatter(reducer.embedding_[:, 0], reducer.embedding_[:, 1], s=2.5,c='black')
```
