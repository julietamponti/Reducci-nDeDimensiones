```python
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import trimesh
from sklearn import manifold
import umap
from sklearn.manifold import TSNE

mesh = trimesh.load('models/cow.obj')
nodes = mesh.vertices
decimacion = nodes.shape[0]//3000
if decimacion==0:
    decimacion=1
nodes = np.array([nodes[i] for i in range(nodes.shape[0]) if i%decimacion==0])

x = nodes[:,0]
y = nodes[:,2]
z = nodes[:,1]

fig = plt.figure(dpi=150)
ax = fig.gca(projection='3d')
ax.scatter(x, y, z)
ax.view_init(elev=10., azim=0)

plt.show()
```

```python
fig = plt.figure(dpi=150)
ax = fig.gca(projection='3d')
ax.scatter(x, y, z)
ax.view_init(elev=90., azim=0)

plt.show()
```

```python
fig = plt.figure(dpi=150)
ax = fig.gca(projection='3d')
ax.scatter(x, y, z)
ax.view_init(elev=0, azim=90)

plt.show()
```

# PCA

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
principalComponents = pca.fit_transform(nodes)
plt.scatter(principalComponents[:,0],principalComponents[:,1])
```

# ISOMAP

```python
# 2 VECINOS
iso = manifold.Isomap(n_neighbors=2, n_components=2)
iso.fit(nodes)
manifold_2da = iso.transform(nodes)
plt.scatter(manifold_2da[:,0],manifold_2da[:,1])
```
