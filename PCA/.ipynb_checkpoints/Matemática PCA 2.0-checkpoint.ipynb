## Matemática de PCA

```python
import warnings
warnings.simplefilter("ignore")
import numpy as np
import mglearn
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

Genero una nube de datos aleatorios utilizando numpy para aplicar PCA

```python
x1 = np.linspace(-10, 10, 100)

# Ahora se genera una relación lineal, pero con más ruido
x2 = np.array([xi*2 + np.random.normal(loc=0, scale=0.5) for xi in x1]) 

X = np.matrix(list(zip(x1, x2)))
```

Graficamos los datos para obtener una idea de que se trata:

```python
sns.set_style('darkgrid')
plt.scatter(x1, x2);
plt.title("Grafico de la nube de datos")
```

Aplicamos la herramienta de SKLearn de PCA

```python
from sklearn.decomposition import PCA

pca = PCA()
transformed = pca.fit_transform(X)
```

```python
# nos devuelve los ejes principales, que representan las direcciones de máxima variación en los datos.
pca.components_ 
```

#### Interpretación de los resultados

* PCA transforma el conjunto de datos a lo largo de los ejes principales.
* El primero de estos ejes está diseñado para capturar la máxima variacion dentro de los datos
    * A partir de esto, se construyen los ejes adicionales que son ortognoales a los ejes de mayor varianza pero que sigen tenienod en cuneta la idea de capturar la mayor cantidad de varianza restante posible.

Para el caso actual, los ejes en los que se proyectaron los datos se ven así:

```python
plt.scatter(x1, x2);
ax1, ay1 = pca.mean_[0], pca.mean_[1]
ax2, ay2 = pca.mean_[0] + pca.components_[0][0], pca.mean_[1] + pca.components_[0][1]
ax3, ay3 = pca.mean_[0] + pca.components_[1][0], pca.mean_[1] + pca.components_[1][1]
plt.plot([ax1, ax2], [ay1, ay2], color='red')
plt.plot([ax2, ax3], [ay2, ay3], color='red');
plt.title("Ejes de las componentes principales")
```

Graficamos entonces el mismo conjunto de datos, pero rotado sobre los ejes rojos propuestos:

```python
plt.scatter(transformed[:,0], transformed[:,1])
plt.axhline(color='red')
plt.axvline(color='red');
plt.title("Datos sobre los ejes de las componentes principales")
```

Debemos observar en este caso, que es como si estuvieramos haciendo zoom sobre la mayor concentracion de la nube de datos, es por eso que los valores del eje *y* han disminuido considerablemente.

Ahora volvemos a graficar con los valores del eje *y* original.

```python
plt.scatter(transformed[:,0], transformed[:,1])
plt.axhline(color='red')
plt.axvline(color='red')
plt.ylim(-10,10);
plt.title("Datos sobre los ejes de las componentes principales")
```

Entonces, la interpretación de lo sucedido es la siguiente:

![interpretacion pca](interpretación_pca.png)

* Si bien PCA se usa para reducir realmente el número de dimensiones.
    * En este caso, simplemente se uso para parametrizar el conjunto de datos a lo largo de nuevos ejes.
