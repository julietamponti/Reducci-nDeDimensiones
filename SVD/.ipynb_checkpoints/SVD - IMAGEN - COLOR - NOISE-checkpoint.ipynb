# SVD

## Trabajando con una imagen a color

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import time

from PIL import Image
```

```python
img = Image.open('fiuba.jpg')

imggray = img.convert()
plt.figure(figsize=(12, 6))
plt.imshow(imggray)
```

```python
#convierto la imagen en una RGB numpy matrix
coloredImg = np.array(img)

plt.figure(figsize=(12,6))
plt.imshow(coloredImg);
```

```python
# Se divide en rojo, verde y azul
R = coloredImg[:, :, 0].copy()
G = coloredImg[:, :, 1].copy()
B = coloredImg[:, :, 2].copy()


plt.figure(figsize=(12,6))
plt.imshow(R);

plt.figure(figsize=(12,6))
plt.imshow(G);


plt.figure(figsize=(12,6))
plt.imshow(B);
```
