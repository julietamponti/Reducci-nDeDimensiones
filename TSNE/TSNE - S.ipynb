## TSNE

```python
import warnings
warnings.simplefilter("ignore")
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
import mpl_toolkits.mplot3d
from sklearn import manifold, datasets
```

```python
original_X, color = datasets.make_s_curve(1000, random_state=0)
```

```python
# plot
plt.rcParams["font.size"] = 18
fig = plt.figure(figsize=(7, 6))
ax = fig.add_subplot(111, projection='3d')
ax.set_xlabel("x1")
ax.set_ylabel("x2")
ax.set_zlabel("x3")
p = ax.scatter(original_X[:, 0], original_X[:, 1], original_X[:, 2], c=color)
plt.tight_layout()
plt.show()
```

```python
autoscaled_X = (original_X - original_X.mean(axis=0))
```

```python
Z = TSNE(perplexity=85, n_components=2, init='pca',
         random_state=100).fit_transform(autoscaled_X)
```

```python
# plot despus de aplicar tSNE
plt.figure(figsize=(6, 6))
plt.scatter(Z[:, 0], Z[:, 1], c=color)
plt.xlabel("z1")
plt.ylabel("z2")
plt.tight_layout()
plt.show()
```
