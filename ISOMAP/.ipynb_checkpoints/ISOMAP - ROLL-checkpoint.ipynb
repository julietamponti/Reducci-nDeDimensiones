```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from sklearn.neighbors import NearestNeighbors
```

## ISOMAP ROLL

```python
# Generamos los datos de forma random
n = 1000
x = np.random.rand(2,n)

# Transformamos los datos en un rollo
v = 3*np.pi/2*(.1 + 2*x[0,:])
X  = np.zeros([3,n])
X[1,:] = 20*x[1,:]
X[0,:] = - np.cos(v)*v
X[2,:] = np.sin(v)*v

from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(14,10))
ax = fig.add_subplot(111, projection="3d")
ax.scatter(X[0,:], X[1,:], X[2,:], c=plt.cm.jet((X[0,:]**2+X[2,:]**2)/100), s=200, lw=0, alpha=1)
ax.set_xlim(np.min(X[0,:]),np.max(X[0,:]))
ax.set_ylim(np.min(X[1,:]),np.max(X[1,:]))
ax.set_zlim(np.min(X[2,:]),np.max(X[2,:]))
ax.axis("off");
```

```python
from sklearn.metrics import pairwise_distances
# Se calcula la matriz de distancias a partir de la matriz X
dist_matrix = pairwise_distances(X)
dist_matrix.shape

# X es la matriz de datos (que contiene la cantidad de muestras y los features) y k la cantidad de vecinos
def nearest_neighbors(X, k): 
    knn_matrix = np.zeros((len(X), k))
    # Se realiza el calculo de distancias
    dist_matrix = pairwise_distances(X)
    # Para cada fila de la matriz X, se van a encontrar los K vecinos mas cercanos
    for i in range(len(X)):
        knn_matrix[i] = dist_matrix[i,:].argsort()[1:k+1]
    return knn_matrix
```

```python
# El formato de la matriz X es (n_samples, n_features)
X = np.transpose(X)

fig = plt.figure(figsize=(14,10))
ax = fig.add_subplot(111, projection="3d")

ax.scatter(X[:,0], X[:,1], X[:,2], c=plt.cm.jet((X[:,0]**2+X[:,2]**2)/100), s=200, lw=0, alpha=1);

# Para cada dato se van a trazar lineas que lo conecten con los vecinos mas cercanos
k = 6 # Cantidad de vecino más cercanos que queremos contemplar
knn = nearest_neighbors(X, k)
for i in range(len(X)):
    neighbors = knn[i]
    for j in range(len(neighbors)):
        ax.plot(X[[i, neighbors.astype('int')[j]], 0], 
                X[[i, neighbors.astype('int')[j]], 1], 
                X[[i, neighbors.astype('int')[j]], 2], color='black');


ax.axis("off")
ax.set_xlim(np.min(X[:,0]),np.max(X[:,0]))
ax.set_ylim(np.min(X[:,1]),np.max(X[:,1]))
ax.set_zlim(np.min(X[:,2]),np.max(X[:,2]))
plt.show();
```

```python
from sklearn.manifold import Isomap
data = X

# Se aplica Isompa con dos componentes (dos dimensiones) 
#y considerando los 6 vecinos más cercanos de cada punto
model = Isomap(n_components=2, n_neighbors=6)
proj = model.fit_transform(data)

# Ploteamos la proyeccion generada por Isomap
plt.figure(figsize=(14,10))
plt.scatter(proj[:, 0], proj[:, 1], c=plt.cm.jet((X[:,0]**2+X[:,2]**2)/100), s=200, lw=0, alpha=1)

# Se trazan las lineas que conectan a cada dato con 
#sus 6 vecinos más cercanos
for i in range(len(X)):
    neighbors = knn[i]
    for j in range(len(neighbors)):
        plt.plot(proj[[i, neighbors.astype('int')[j]], 0], 
                 proj[[i, neighbors.astype('int')[j]], 1], color='black');
        
plt.title('Isomap con k-Nearest Neighbors = ' + str(k), size=30)
plt.axis("off")
plt.show();
```
